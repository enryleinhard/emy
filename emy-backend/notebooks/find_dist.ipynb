```python
res_df = None

for x in range(file_count):
    
    print(f'{x}_loop')

    origins = []
    destinations = []
    for i in range(0,10):
        origins.append(str(dataframes_arr[x][i][2]) + ' ' + str(dataframes_arr[x][i][3])) #get origin lat long
        destinations.append(str(dataframes_arr[x][i][5]) + ' ' + str(dataframes_arr[x][i][6])) #get dest lat long

    matrix = gmaps.distance_matrix(origins, destinations, mode="driving")
    matrixdf = json_normalize(matrix,['rows','elements'])
    
    temp_df = pd.DataFrame(dataframes_arr[x], columns = ['index', 'orig_store','orig_lat','orig_long','dest_store','dest_lat','dest_long'])
    temp_df = temp_df.reset_index()  # make sure indexes pair with number of rows

    matrixdf['orig_store'] = temp_df['orig_store']
    matrixdf['orig_lat'] = temp_df['orig_lat']
    matrixdf['orig_long'] = temp_df['orig_long']
    matrixdf['dest_store'] = temp_df['dest_store']
    matrixdf['dest_lat'] = temp_df['dest_lat']
    matrixdf['dest_long'] = temp_df['dest_long']
    matrixdf = matrixdf.head(10)

    if res_df is None:
        res_df = matrixdf.copy()
    else:
        res_df = pd.concat([res_df, matrixdf], axis=0, ignore_index=True)
    
res_df
```
